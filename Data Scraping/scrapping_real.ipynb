```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_banks_in_Nepal'

page = requests.get(url)

soup = BeautifulSoup(page.text, 'html')
print(soup)
```

```python
<table class="wikitable sortable static-row-numbers static-row-header-text jquery-tablesorter">
```

```python
table = soup.find_all('table')[0]
print(table)
```

```python
titles = table.find_all('th')
print(titles)
```

```python
world_title = [title.text.strip() for title in titles]
print(world_title)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(columns=world_title)
df
```

```python
column_data = table.find_all('tr')
```

```python
for row in column_data[1:]:
    row_data = row.find_all('td')
    ind_row_data = [data.text.strip() for data in row_data]
    length  = len(df)
    df.loc[length] = ind_row_data
```

```python
print(df.head(10))
```

```python
print(df.to_string(index=False))
```

```python
df.columns = ['Bank', 'Location', 'Capital', 'Assets', 'Branches', 'Website']
```

```python
print(df.to_string(index=False))
```
